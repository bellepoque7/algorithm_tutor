- immutable 객체 int를 변경해보면 주소가 바뀐다

```python
a = 1
#a가 내부적으로 어떤 객체를 가르키는지 확인
print(id(a))
a = 2
print(id(a))
```

```python
#마치 deep copy 처럼 작동함. (mutable의 경우)
b = a
# 값이 똑같아서 같은 객체를 알려줌
print(id(a),id(b),id(2))
```

```python
# 값이 달라지면 id 값이 달라질까? 
a = 5
print(id(a))
print(id(b))
#정답: 네 
```

- mutable 객체 list를 변경하면 주소가바뀌지만, append 하면 동일하다.

```python
c = [1,2,3]
print(id(c))
c = [4,5,6]
print(id(c))
c.append(7)
print(id(c))
```

```python
# 할당하면 int(mutable) 처럼 작동할까?
d = c
d.append(8)
print(d)
print(c)
# mutable 객체는 이래서 deep copy 가 필요하다. 
```

```python
import copy
e = copy.deepcopy(c)
print(c,e)
e.append(9)
print(c,e)
```

결론. dictionary 의 key는 immutable 객체만 가능하다.
